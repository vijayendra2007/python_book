```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=[1,2,3,4,5,]
y=[1,4,9,16,25]

#line plot 
plt.plot(x,y,marker ='o',markersize=12,color='blue',linestyle='-.')
plt.xlabel('x values ')
plt.ylabel('y values ')
plt.title('lineplot')
plt.show()
```

```python

sns.scatterplot(x="body_mass_g",y="bill_depth_mm",hue="sex",data=data)
plt.show()
```

```python
Categories=['bill_length_mm','bill_depth_mm','flipper_length_mm',]
values=[40.3,18.0,195.0,]
color=['blue','green','red']
plt.bar(Categories,values,color=color)
plt.title('Bar Plot')
plt.xlabel('penguin data')
plt.ylabel('values')
plt.show()
```

```python
Categories=['bill_length_mm','bill_depth_mm','flipper_length_mm',]
values=[40.3,18.0,195.0]
explode=(0,0.2,0,)

plt.pie(values, explode=explode ,labels=Categories,autopct="%1.1f%%",shadow=True)
```

```python
import numpy as np
corr=data.drop(columns="month").corr()
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]
sizes = [100, 200, 300, 400, 500]  
plt.scatter(x, y, s=sizes, alpha=1, c='cyan', edgecolors='black', linewidth=0.5)
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Bubble Chart ')
plt.show()
```
